# Gaussian Noise

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

## get inputs

```python
# read input
image = Image.open('data/input.jpg').convert('LA')
```

```python
# convert inputs
image_arr = np.asarray(image)
image_arr = image_arr[:,:,0]
```

```python
# show input
plt.imshow(image_arr, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## add gaussian noise

```python
def add_gaussian(image_arr):
    mean = np.mean(image_arr)
    var = np.var(image_arr)
    # mean = 0
    # var = 0.1
    sigma = var ** 0.5
    gauss = np.random.normal(mean, sigma, image_arr.shape)
    res = gauss + image_arr
    res = np.clip(res, 0, 255).astype('uint8')
    return res
```

```python
image_with_noise = add_gaussian(image_arr)
```

```python
plt.imshow(image_with_noise, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
cv2.imwrite('result/GaussianNoise.png',image_with_noise)
```

## Median Filter

```python
med_flt = cv2.medianBlur(image_with_noise, 3)
```

```python
plt.imshow(med_flt, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
cv2.imwrite('result/MedianRes.png',med_flt)
```

## Gaussian Filter

```python
gus_flt = cv2.GaussianBlur(image_arr, (5,5), 0)
```

```python
plt.imshow(gus_flt, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
cv2.imwrite('result/GaussianRes.png',gus_flt)
```
